```python
nr_bins = int(1e+6)
threshold = int(10)
csv_path = '../cache/train_X.csv'
gbdt_path = '../cache/tr.gbdt.out'
out_path = '../cache/tr.ffm'
```

```python
def gen_hashed_fm_feats(feats, nr_bins):
    feats = ['{0}:{1}:1'.format(field-1, hashstr(feat, nr_bins)) for (field, feat) in feats]
    return feats
```

```python
frequent_feats = read_freqent_feats(threshold)
```

```python
frequent_feats
```

```python
with open('../cache/tr.ffm', 'w') as f:
    print('writing ffm..\n')
    for row, line_gbdt in tqdm(zip(csv.DictReader(open('../cache/train_X.csv')),
                                   open('../cache/tr.gbdt.out'))):
        y = '0'
        label='target'
        if label in row:
             y = row[label]
             del row[label]

        feats = []

        l = gen_feats2(row)
        for i, feat in enumerate(l):
            
            field = feat.split('-')[0]
            
            # t, field = field[0], int(field[1:])
            t = (True if '_cat' in field else False)
            if t == True and feat not in frequent_feats:
                feat = feat.split('-')[0]+'less'
            #if t == True:
            field = i
            feats.append((field, feat))

        for i, feat in enumerate(line_gbdt.strip().split()[1:], start=1):
            field = i + 39
            feats.append((field, str(i)+":"+feat))

        #print(feats)
        feats = gen_hashed_fm_feats(feats, nr_bins)
        # f.write(row['Label'] + ' ' + ' '.join(feats) + '\n')
        f.write(y + ' ' + ' '.join(feats) + '\n')
```
